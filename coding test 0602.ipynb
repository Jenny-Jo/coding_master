```python
def Meantri_rmv(arr):
    arr.sort()
    low5_idx= int(len(arr)*0.05)
    high5_idx=int(len(arr)*0.95)
    
    arr = arr[low5_idx : high5_idx]

    mean = sum(arr)/int(len(arr))        

    return mean
```

```python
def bank(n):
    if n < 7 :
        ans = sum([i+1 for i in range(n)])
    else:
        a, b = divmod(n,7)
        sum1=sum([i+1 for i in range(7)])
        ans = a*sum1 + sum([7*i for i in range(a)])+sum([i+a for i in range(b)])
    return ans
        
        
```

```python
def bank_1(n):
    if n<7:
        ans = sum([i+1 for i in range(n)])
    else:
        a, b = divmod(n,7)
        ans=0
        
        div = [ i+a+1 for a in range(a) for i in range(7)]
        mod = [ i+a+1 for i in range(b)]
        ans += sum(div) + sum(mod)
                         
    return ans
            
```
